### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.arange(1, 6)
y = x**2

# Bar chart data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Scatter plot data
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

# Pie chart data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create the subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Plot 1: Line chart
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')

# Plot 2: Bar chart
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Plot 3: Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='purple')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')

# Plot 4: Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
