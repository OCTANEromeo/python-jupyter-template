### Exercise 1: Line Plot Customization

```python
# Import necessary libraries
import matplotlib.pyplot as plt  # For plotting
import numpy as np               # For handling numbers

# Step 1: Create x and y values
x = np.arange(0, 10.1, 0.1)   # x values from 0 to 10 with a step of 0.1
y = x**2                      # y values are the square of x

# Step 2: Create the plot
plt.plot(x, y, color='blue')   # Plot x vs. y with a blue line

# Step 3: Add labels and title
plt.title('Square Function: y = x^2')   # Title of the plot
plt.xlabel('X values')                  # Label for the x-axis
plt.ylabel('Y values')                  # Label for the y-axis

# Step 4: Add grid and show the plot
plt.grid(True)   # Display grid lines
plt.show()       
```
